# **AI model**

## Training

### Monitered solo training

```python
dataset = fluoride
name = 'fluoride'
train_x = dataset[['Temperature (°C)', 'Pressure (bar)', 'pH Level', 'Brine Concentration (g/L)', 'Flow Rate (L/min)', 'Energy Input (kWh)', 'Water Recovery (%)', 'Evaporation Rate (L/h)', 'Residence Time (min)']]
train_y = dataset[['Amount Extracted (mg)']]
model = tf.keras.models.Sequential([
    tf.keras.layers.Dense(32, activation='relu', input_shape=(9,)),
    tf.keras.layers.Dense(1, activation='relu')
])

my_optimizer = tf.keras.optimizers.Adam(learning_rate=0.001)
model.compile(optimizer=my_optimizer, loss=tf.keras.losses.Huber())
history = model.fit(train_x, train_y, epochs=200, validation_split=0.2)

loss = history.history['loss']
val_loss = history.history['val_loss']
plt.figure(figsize=(10, 6))

# Plot training loss
plt.plot(loss, label='Training Loss')

# Plot validation loss
plt.plot(val_loss, label='Validation Loss')

# Add labels and title
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title(f'Training and Validation Loss ({name})')
plt.legend()
plt.grid(True)

plt.savefig(f'{name}.png')
```

```python
model.save(f'/content/models/{name}.h5')
```

### Unmonitered group training

```python
def compile_model(train_x, train_y, name):
  model = tf.keras.models.Sequential([
    tf.keras.layers.Dense(16, activation='relu', input_shape=(9,)),

    tf.keras.layers.Dense(1, activation='relu')
  ])

  my_optimizer = tf.keras.optimizers.Adam(learning_rate=0.1)
  model.compile(optimizer=my_optimizer, loss=tf.keras.losses.Huber())
  history = model.fit(train_x, train_y, epochs=100, validation_split=0.2)

  model.save(f'/content/models/{name}.h5')

  loss = history.history['loss']
  val_loss = history.history['val_loss']
  plt.figure(figsize=(10, 6))

  # Plot training loss
  plt.plot(loss, label='Training Loss')

  # Plot validation loss
  plt.plot(val_loss, label='Validation Loss')

  # Add labels and title
  plt.xlabel('Epochs')
  plt.ylabel('Loss')
  plt.title(f'Training and Validation Loss ({name})')
  plt.legend()
  plt.grid(True)

  # Saving the figure
  plt.savefig(f'{name}.png')
```

```python
for i in range(len(mineral_list)):
  dataset = mineral_list[i]
  name = mineral_names[i]
  train_x = dataset[['Temperature (°C)', 'Pressure (bar)', 'pH Level', 'Brine Concentration (g/L)', 'Flow Rate (L/min)', 'Energy Input (kWh)', 'Water Recovery (%)', 'Evaporation Rate (L/h)', 'Residence Time (min)']]
  train_y = dataset[['Amount Extracted (mg)']]
  compile_model(train_x, train_y, name)
```

# **Generating best case scenario for maximum sodium extraction**

```python
def find_best_case(model, scaler, pre_list, num_iterations=100):
  best_conditions = None
  best_output = -float('inf')  # We want to maximize sodium extraction
  for _ in range(num_iterations):
      # Simulate random conditions
      temp = random.uniform(70, 100)
      pressure = random.uniform(1.0, 2.5)
      ph = random.uniform(7.0, 9.0)
      brine_concentration = random.uniform(90, 150)
      flow_rate = random.uniform(1.0, 3.5)
      energy_input = random.uniform(70, 100)
      water_recovery = random.uniform(70, 95)
      evaporation_rate = random.uniform(1.0, 2.0)
      residence_time = random.uniform(40, 80)

      # Generated array
      arr = [temp, pressure, ph, brine_concentration, flow_rate, energy_input, water_recovery, evaporation_rate, residence_time]
      for i, n in enumerate(pre_list):
        if n != 0:
          arr[i] = float(n)
      # Create conditions array
      conditions = np.array([arr])

      # Predict sodium extraction
      output = model.predict(conditions)[0][0]

      # Check if this is the best output so far
      if output > best_output:
          best_output = output
          best_conditions = conditions[0]

      print(f"Running simulation {_} ...")

  return best_conditions, best_output
```
